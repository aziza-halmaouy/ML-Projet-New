```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Charger dataset
df = pd.read_csv("Public_Chauffeurs_20260405.csv")

# colonnes utiles
df = df[['Driver Type', 'License Type', 'Sex', 'Chauffeur City', 'Status']]

# encodage
for col in df.columns:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# RULES 
def rule_based(row):
    if row['Driver Type'] == 1 and row['Sex'] == 1:
        return 1
    elif row['License Type'] == 0:
        return 0
    else:
        return 1

# prédiction règles
df['Predicted'] = df.apply(rule_based, axis=1)

# accuracy simple
accuracy = (df['Predicted'] == df['Status']).mean()

print("Rule-Based Accuracy:", accuracy)
```
